```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fifa = pd.read_csv("fifa_data.csv")

fifa.head(3)
```

```python
fifa.columns
```

```python
fifa.shape
```

```python
# Droping to unneccessary columns
to_drop = ["ID","Unnamed: 0","Photo", "Flag", "Club Logo",'Contract Valid Until']

fifa.drop(columns=to_drop, inplace=True, axis=1)
```

```python
fifa.describe().T
```

```python
corr_matrix = fifa.corr(numeric_only=True)

if (corr_matrix > .5).sum().sum() > len(fifa.columns):
    plt.figure(figsize=(32, 25))
    sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
    plt.show()
```
