```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales_df = pd.read_csv("./Amazon Sales data.csv")

sales_df.head(3)
```

```python
cols = sales_df.columns.tolist()

cols.insert(0, cols.pop(cols.index("Order ID")))


sales_df = sales_df[cols]
```

```python
sales_df.head(3)
```

```python
sales_df.info()
```

```python
sales_df["Units Sold"] = sales_df["Units Sold"].astype(float)
```

```python
sales_df.info()
```

```python
sales_df.shape
```

```python
sales_df.isnull().sum().sort_values(ascending=False)
```

```python
sales_df.describe().T
```

```python
sales_df["Ship Date"] = pd.to_datetime(sales_df["Ship Date"], format='mixed')
```

```python
sales_df["Order Date"] = pd.to_datetime(sales_df["Order Date"], format="mixed")
```

```python
# Data Analysis:

# Queries:

# Which regions have the highest total sales revenue?

# What is the average unit price and unit cost for each item type?

# Which country has the highest total profit?

# How does the sales channel affect the order priority distribution?

# What is the average order processing time (duration between order and ship dates) for each sales channel?

# Which item types have the highest and lowest total sales?

# How does the order priority vary across different regions?

# What is the correlation between unit price and total profit?

# Are there any seasonal trends or patterns in the sales data?

# How does the number of units sold vary across different countries?
```

```python
sales_df.columns
```

```python
# 1.Which regions have the highest total sales revenue?
highest_revenue_regions = sales_df.groupby(sales_df["Region"])["Total Revenue"].sum().sort_values(ascending=False)
highest_revenue_regions.idxmax()
```

```python
plt.figure(figsize=(15, 5))
sns.set_style('darkgrid')
sns.barplot(x=highest_revenue_regions.index, y=highest_revenue_regions)
plt.xticks(rotation= 90)
plt.title('Highest Total Sales Revenue by Region', fontsize= 20, color= 'Blue')
plt.xlabel('Regions', fontsize= 20, color= 'red')
plt.ylabel('Revenue', fontsize= 20, color= 'c')
plt.show()
```

```python
#2.What is the average unit price and unit cost for each item type?

average_unit_price = sales_df.groupby(sales_df["Item Type"])["Unit Price"].mean().sort_values(ascending=False)
average_unit_cost = sales_df.groupby(sales_df["Item Type"])["Unit Cost"].mean().sort_values(ascending=False)

average_unit_data = pd.DataFrame({
    "Average Unit Price": average_unit_price,
    "Average Unit Cost":average_unit_cost
})

average_unit_data
```

```python
plt.figure(figsize=(15, 5))

sns.barplot(x=average_unit_price.index, y=average_unit_price)
plt.xticks(rotation= 90)
plt.title('Average Unit Price per Item Type')
plt.xlabel('Item Type')
plt.ylabel('Unit Price')
plt.show()
```

```python
plt.figure(figsize=(15, 5))

sns.barplot(x=average_unit_cost.index, y=average_unit_cost)
plt.xticks(rotation= 90)
plt.title('Average Unit Cost per Item Type')
plt.xlabel('Item Type')
plt.ylabel('Unit Cost')
plt.show()
```

```python
sales_df.columns
```

```python
#3.Which country has the highest total profit?

highest_total_profit_countries = sales_df.groupby(sales_df["Country"])["Total Profit"].sum()

highest_total_profit_countries.idxmax()
```

```python
plt.figure(figsize=(15, 5))
sns.set_style('darkgrid')
sns.barplot(x=highest_total_profit_countries.index, y=highest_total_profit_countries)
plt.xticks(rotation=90)
plt.title("Highest Total Profit Countries")
plt.xlabel("Country")
plt.ylabel("Total Profit")
plt.show()
```

```python
#4.How does the sales channel affect the order priority distribution?

Sales_Channel_Order_Priority_Distribution = sales_df.groupby(sales_df["Sales Channel"])["Order Priority"].value_counts()
```

```python
Sales_Channel_Order_Priority_Distribution = sales_df.groupby(['Sales Channel', 'Order Priority'])["Order Priority"].count()
Sales_Channel_Order_Priority_Distribution = Sales_Channel_Order_Priority_Distribution.reset_index(name="Count")

plt.figure(figsize=(10, 5))
sns.barplot(x="Sales Channel", y="Count",hue='Order Priority',data=Sales_Channel_Order_Priority_Distribution)
plt.show()
```

```python
# 5.What is the average order processing time (duration between order and ship dates) for each sales channel?

sales_df["Processing Time"] = sales_df["Ship Date"] -sales_df["Order Date"] 
average_processing_time = sales_df.groupby(sales_df["Sales Channel"])["Processing Time"].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=sales_df["Sales Channel"], y=sales_df["Processing Time"].dt.days, data=sales_df)
plt.show()
```

```python
sales_df.columns
```

```python
#6. Which item types have the highest and lowest total sales?

grouped_items = sales_df.groupby(sales_df["Item Type"])["Total Revenue"].sum()

print(grouped_items.idxmax())
print(grouped_items.idxmin())

```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=grouped_items.index, y=grouped_items)
max_index = grouped_items.idxmax()
plt.scatter(x=max_index, y=grouped_items[max_index], color="Green")

min_index = grouped_items.idxmin()
plt.scatter(x=min_index, y=grouped_items[min_index], color="Red")
plt.show()
```

```python
#7.How does the order priority vary across different regions?

order_priority_across_regions = sales_df.groupby(sales_df["Order Priority"])["Region"].value_counts()
order_priority_across_regions
```

```python
order_priority_across_regions = sales_df.groupby(sales_df["Order Priority"])["Region"].value_counts().reset_index(name="Order Count")
plt.figure(figsize=(15, 5))
sns.barplot(x="Region", y="Order Count",hue="Order Priority", data=order_priority_across_regions)
plt.show()
```

```python
#8. What is the correlation between unit price and total profit?

price_corr_with_profit = sales_df["Units Sold"].corr(sales_df["Total Profit"])
print(f"Corralation between Units Sold and Total Profit: {price_corr_with_profit}")
```

```python
plt.figure(figsize=(4,2))
plt.scatter(x= price_corr_with_profit, y= price_corr_with_profit, s= 200, color= 'RED' )
plt.xticks(np.arange(-1,2,0.5))
plt.yticks(np.arange(-1,2,0.5))
plt.title('Correlation Between Units Price and Total Profits')

plt.show
```

```python
# Are there any seasonal trends or patterns in the sales data?

month_names = {
    1:"Jan",
    2 :"Feb",
    3 : "Mar",
    4 :"Apr",
    5:"May",
    6: "June",
    7:"July", 
    8:"Aug",
    9:"Sep",
    10: "Oct",
    11:"Nov",
    12:"Dec"
}

month_sales = sales_df.groupby(sales_df["Order Date"].dt.month)["Total Revenue"].sum()
month_sales.index = month_sales.index.map(month_names)

month_sales


plt.figure(figsize=(15, 5))
sns.barplot(x=month_sales.index , y=month_sales)
plt.show()
```

```python
# 10.How does the number of units sold vary across different countries?

units_sold_in_countries = sales_df.groupby(sales_df["Country"])["Units Sold"].sum()

plt.figure(figsize=(15, 5))
plt.xticks(rotation=90)
sns.barplot(x=units_sold_in_countries.index, y=units_sold_in_countries)
plt.show()
```

```python
# 11.How does the total sales revenue vary across different countries?

total_revenue_countries = sales_df.groupby(sales_df["Country"])["Total Revenue"].sum()
plt.figure(figsize=(18, 7))
sns.barplot(x=total_revenue_countries.index, y=total_revenue_countries)
plt.xticks(rotation=90)
plt.show()
```

```python
# 12.What is the distribution of unit prices for each item type?

distribution_per_item = sales_df.groupby(sales_df["Item Type"])["Unit Price"].sum().reset_index("Item Type")
plt.pie(x=distribution_per_item["Unit Price"], labels=distribution_per_item["Item Type"], autopct='%1.1f%%')
plt.show()
```

```python
# 13.Which sales channel has the highest average unit price?

highest_unit_price_channels = sales_df.groupby(sales_df["Sales Channel"])["Unit Price"].mean().reset_index(name="Unit Price")

plt.pie(x=highest_unit_price_channels["Unit Price"], labels=highest_unit_price_channels["Sales Channel"], autopct='%1.1f%%')
plt.title("Average Unit Price per Sales Channel", color="Red")
plt.show()
```

```python
# 14.Are there any outliers in the total cost distribution?

q1= sales_df['Total Cost'].quantile(0.25)
q3= sales_df['Total Cost'].quantile(0.75)

iqr= q3-q1

lower_fence= q1-1.5*iqr
upper_fence= q3+1.5*iqr

outliers= sales_df[(sales_df['Total Cost']<lower_fence)| (sales_df['Total Cost']>upper_fence)].reset_index(drop= True)
outliers
```

```python
# 15.How does the total profit vary across different item types?

highest_total_profit_item_types = sales_df.groupby(sales_df["Item Type"])["Total Revenue"].sum().reset_index(name="Revenue")

plt.pie(x=highest_total_profit_item_types["Revenue"], labels=highest_total_profit_item_types["Item Type"], autopct='%1.1f%%')
plt.axis("equal")
plt.show()
```

```python
# 16.What is the average order processing time for each country?

avg_ordering_time_per_country = sales_df.groupby(sales_df["Country"])["Processing Time"].mean()

df_ordering = pd.DataFrame(
   avg_ordering_time_per_country)

df_ordering
```

```python
# 17.Which region has the highest average total revenue per order?

sales_df["Average Revenue Per Order"] = sales_df["Total Revenue"]/sales_df["Units Sold"]

highest_revenue_regions = sales_df.groupby(sales_df["Region"])["Average Revenue Per Order"].mean().reset_index(name="Revenue")
highest_revenue_regions

plt.pie(x=highest_revenue_regions["Revenue"],labels=highest_revenue_regions["Region"], autopct='%1.1f%%')
plt.show()
```

```python
# 18.Is there a relationship between the number of units sold and the total profit?


corr_between_unit_sold_total_profit = sales_df["Units Sold"].corr(sales_df["Total Profit"])
print(f"Relationship between Units Sold and Total Revenue: {corr_between_unit_sold_total_profit}")
```

```python
# 19.How does the order priority vary based on the item type?

order_item_priotity = sales_df.groupby(sales_df["Item Type"])["Order Priority"].value_counts().reset_index(name="Count")

plt.figure(figsize=(18, 8))
sns.barplot(data=order_item_priotity, x=order_item_priotity["Item Type"], y=order_item_priotity["Count"], hue="Order Priority")
plt.show()
```
