```python
# Checking for how often a customer ordered

customer_order_frequency = super_store.groupby(["Customer Name","Customer ID","Segment"])["Order ID"].count().reset_index(name="Total Orders")

# customer_order_frequency.rename(columns={"Order ID":"Total Orders"}, inplace=True)

top_10_most_freq_ordered_customers = customer_order_frequency.sort_values(by="Total Orders", ascending=False).head(10)
top_10_most_freq_ordered_customers
```

```python
customer_sales = super_store.groupby(["Customer Name","Customer ID","Segment"])["Sales"].sum().reset_index(name="Total Amount")

top_spenders = customer_sales.sort_values(by="Total Amount", ascending=False).head(12)

top_spenders
```

```python
high_profited_category_goods = super_store.groupby(super_store["Category"])["Profit"].sum()

plt.pie(x=high_profited_category_goods, labels=high_profited_category_goods.index, autopct='%1.1f%%')
plt.show()
```

```python
dist_sub_category_goods = super_store.groupby(super_store["Sub-Category"])["Profit"].sum()

plt.figure(figsize=(15, 8))
sns.barplot(x=dist_sub_category_goods.index, y=dist_sub_category_goods)
plt.title("Profit brought by goods")
plt.show()
```

```python
dist_sub_category_goods = super_store.groupby(super_store["Sub-Category"])["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(15, 8))
sns.barplot(y=dist_sub_category_goods.index, x=dist_sub_category_goods)
plt.title("Distrubution of sales by sub categories")
plt.show()
```

```python
super_store["Order Date"] = pd.to_datetime(super_store["Order Date"])

super_store["Order Date"].describe()
```

```python
yearly_sales = super_store.groupby(super_store["Order Date"].dt.year)["Sales"].sum()

sns.lineplot(x=yearly_sales.index, y=yearly_sales,marker='o', linestyle='-')
plt.title("Yearly Sales Distribution")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.show()
```

```python
yearly_profit = super_store.groupby(super_store["Order Date"].dt.year)["Profit"].sum()

sns.lineplot(x=yearly_profit.index, y=yearly_profit, marker='o', linestyle='-')
plt.title("Yearly Sales Distribution")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.show()
```

```python
year_sales = super_store.loc[super_store["Order Date"].dt.year == 2016]
quarter_sales = year_sales.resample("Q", on="Order Date")["Sales"].sum()

plt.figure(figsize=(15, 8))
sns.lineplot(x=quarter_sales.index, y=quarter_sales, marker="o", color="Red")
plt.show()
```

```python
year_profit = super_store.loc[super_store["Order Date"].dt.year == 2016]

quarter_profit = year_profit.resample("Q", on="Order Date")["Profit"].sum()

plt.figure(figsize=(15, 8))
sns.lineplot(x=quarter_profit.index, y=quarter_profit, marker="o", linestyle="--")
plt.title("Profit by quarter")
plt.xlabel("Year Quarter")
plt.ylabel("Amount")
plt.show()
```

```python
year_2016_profit = super_store.loc[super_store["Order Date"].dt.year == 2016]
mothly_profit = year_2016_profit.resample("M", on="Order Date")["Profit"].sum()

plt.figure(figsize=(15, 8))
sns.lineplot(x=mothly_profit.index, y=mothly_profit, marker="o", linestyle="--", color="Orange")
plt.title("Profit by quarter")
plt.xlabel("Year Quarter")
plt.ylabel("Amount")
plt.show()
```

```python
years = [2014, 2015, 2016, 2017]

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

axes = axes.flatten()


for i, year in enumerate(years):
   
    year_sales = super_store.loc[super_store["Order Date"].dt.year == year]

    monthly_sales = year_sales.resample("M", on="Order Date")["Sales"].sum()


    sns.lineplot(x=monthly_sales.index, y=monthly_sales, marker="o", linestyle="--", ax=axes[i])
    axes[i].set_title(f'Sales for {year}')


plt.tight_layout()
plt.show()
```

```python
years = [2014, 2015, 2016, 2017]


fig, axes = plt.subplots(2, 2, figsize=(15, 10))

axes = axes.flatten()

for i, year in enumerate(years):

    year_profitt = super_store.loc[super_store["Order Date"].dt.year == year]

    monthly_profitt = year_profitt.resample("M", on="Order Date")["Profit"].sum()

    sns.lineplot(x=monthly_profitt.index, y=monthly_profitt, marker="o", linestyle="--", ax=axes[i])
    axes[i].set_title(f"Profits in {year}")

plt.tight_layout()
plt.show()
```
